```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
# Input sequence
X = [1, 2, 3]
```

```python
# Initial hidden state
h = 0

# Initial cell state
C = 0
```

```python
Wf = 0.5
Wi = 0.6
Wc = 0.7
Wo = 0.8

Uf = 0.4
Ui = 0.3
Uc = 0.2
Uo = 0.1

bf = 0.1
bi = 0.1
bc = 0.1
bo = 0.1
```

```python
for t, x in enumerate(X):

    # Forget gate
    f = sigmoid(Wf*x + Uf*h + bf)

    # Input gate
    i = sigmoid(Wi*x + Ui*h + bi)

    # Candidate cell state
    C_tilde = tanh(Wc*x + Uc*h + bc)

    # Update cell state
    C = f*C + i*C_tilde

    # Output gate
    o = sigmoid(Wo*x + Uo*h + bo)

    # Update hidden state
    h = o * tanh(C)

    print("Time step:", t+1)
    print("Forget gate:", f)
    print("Input gate:", i)
    print("Candidate state:", C_tilde)
    print("Cell state:", C)
    print("Output gate:", o)
    print("Hidden state:", h)
    print("--------------")
```

```python
Wy = 1.2
by = 0.5

y_pred = Wy*h + by

print("Predicted Next Value:", y_pred)
```
